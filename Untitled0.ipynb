```python
import matplotlib.pyplot as plt
import numpy as np

# Données
requests = [10, 100, 500, 1000]
cpu_rest = [8.5, 9.7, 12.1, 14.2]
cpu_soap = [12.1, 13.6, 15.2, 16.3]
cpu_graphql = [6.7, 7.9, 9.4, 11.0]
cpu_grpc = [5.4, 6.2, 7.8, 8.9]

mem_rest = [24.3, 28.5, 35.4, 42.1]
mem_soap = [40.2, 45.0, 55.3, 60.8]
mem_graphql = [18.1, 22.3, 28.7, 35.1]
mem_grpc = [16.8, 19.4, 24.1, 29.7]

# Créer les indices pour les barres
x = np.arange(len(requests))  # Positions des requêtes simultanées
width = 0.2  # Largeur des barres

# Création de la figure et des axes
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Graphique pour la consommation CPU
axs[0].bar(x - 1.5*width, cpu_rest, width, label="REST", color='tab:blue')
axs[0].bar(x - 0.5*width, cpu_soap, width, label="SOAP", color='tab:orange')
axs[0].bar(x + 0.5*width, cpu_graphql, width, label="GraphQL", color='tab:green')
axs[0].bar(x + 1.5*width, cpu_grpc, width, label="gRPC", color='tab:red')

# Paramètres du graphique CPU
axs[0].set_title('Consommation CPU pour différentes charges de travail')
axs[0].set_xlabel('Nombre de Requêtes Simultanées')
axs[0].set_ylabel('CPU (%)')
axs[0].set_xticks(x)
axs[0].set_xticklabels(requests)
axs[0].legend()

# Graphique pour la consommation de mémoire
axs[1].bar(x - 1.5*width, mem_rest, width, label="REST", color='tab:blue')
axs[1].bar(x - 0.5*width, mem_soap, width, label="SOAP", color='tab:orange')
axs[1].bar(x + 0.5*width, mem_graphql, width, label="GraphQL", color='tab:green')
axs[1].bar(x + 1.5*width, mem_grpc, width, label="gRPC", color='tab:red')

# Paramètres du graphique Mémoire
axs[1].set_title('Consommation Mémoire pour différentes charges de travail')
axs[1].set_xlabel('Nombre de Requêtes Simultanées')
axs[1].set_ylabel('Mémoire (MB)')
axs[1].set_xticks(x)
axs[1].set_xticklabels(requests)
axs[1].legend()

# Ajuster la disposition
plt.tight_layout()

# Afficher le graphique
plt.show()
```
